```python
# Importing packages
import pandas as pd
```

```python
# Reading the data
file_path = 'https://raw.githubusercontent.com/owid/covid-19-data/refs/heads/master/public/data/owid-covid-data.csv'
df = pd.read_csv(file_path)
```

```python
# Find the minimum and maximum date
min_date = df['date'].min()
max_date = df['date'].max()

# Print the range
print(f"Date range: {min_date} to {max_date}")
```

```python
## CLEANING & PREPROCESSING FUNCTION

# Function to selected desired columns and remove any NaNs
def cleaning_data(df):

  # Keep only the columns specified by index
  #small_df = df.iloc[:, list(range(0,6)) + [7, 8, 17, 19, 25, 26] + list(range(31,39)) + list(range(49,59)) + [60, 61, 62, 65]]
  small_df = df.iloc[:, list(range(0,6)) + [7, 8, 17, 19, 25, 26] + list(range(31,39)) + [62]]

  # Removing any NaNs
  cleaned_df = small_df.dropna().copy()

  # Convert 'date' column to datetime format
  cleaned_df['date'] = pd.to_datetime(cleaned_df['date'])

  # Return the cleaned dataframe
  return cleaned_df

cleaned_df = cleaning_data(df)
```
